## K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

Since we have two clusters, Private Yes/ No we are creating KMeans with two clusters

```python
k_means = KMeans(n_clusters=2)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

Fitting the model except Private column.

```python
k_means.fit(df.drop('Private', axis=1))
```

** What are the cluster center vectors?**

```python
k_means.cluster_centers_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters



```python
#I'm creating a new column for df called 'Cluster',
# which is a 1 for a Private school, and a 0 for a public school.

def public_or_private(cluster):
    if cluster=='Yes':
        return 1
    elif cluster=='No':
        return 0
```

```python
df['Cluster'] = df['Private'].apply(public_or_private)
# we've created cluster column which has numerical value
```

```python
df
```

```python
# Create a confusion matrix and classification report
# to see how well the Kmeans clustering worked without being given any labels.**
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(df['Cluster'], k_means.labels_ ))
print('\n')
print(classification_report(df['Cluster'], k_means.labels_))
```
